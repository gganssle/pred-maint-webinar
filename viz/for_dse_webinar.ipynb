```python
dat = raw[raw['id'] == 92]

plt.figure(figsize=(15,10))

for i,sensor in enumerate(list(raw.columns)[5:-4]):
    plt.plot(dat[sensor], label=sensor, linestyle='-')

plt.tight_layout()
plt.legend()
#plt.show()
plt.savefig('../dat/sensors_on_92.png', transparent=True, dpi=300)
```
